```python
# 3. Explain the following terms:
# « The difference in np.array, np.asarray and np.asanyarray.
# « The difference between Deep copy and shallow copy.

## np.array creates a new NumPy array from an input object (such as a list, tuple, or another array).

## np.asarray creates an input object to a NumPy array.
## If the input is already a NumPy array of the same data type, no copy is made, and the original array is returned.

## np.asanyarray is similar to np.asarray, but it passes through subclasses of ndarray without converting
## them to a base ndarray.

## Shallow copy creates a new object but inserts references to the objects found in the original,
## so changes to mutable elements will affect both copies.

## Deep copy creates a new object and recursively copies all objects found in the original,
## so changes to any elements do not affect the original.
```

```python
# 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
# origin (quadrants), add a legend, label the axes, and set the title as ‘Quadrant-wise Scatter Plot".

np.random.seed(6)
x = np.random.randn(100)
y = np.random.randn(100)

quadrants = []
for i in range(len(x)):
    if x[i] > 0 and y[i] > 0:
        quadrants.append('Quadrant I')
    elif x[i] < 0 and y[i] > 0:
        quadrants.append('Quadrant II')
    elif x[i] < 0 and y[i] < 0:
        quadrants.append('Quadrant III')
    else:
        quadrants.append('Quadrant IV')

data = {'x': x, 'y': y, 'Quadrant': quadrants}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant')

plt.legend(title='Quadrant')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

plt.show()
```

```python
# 18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
# Wave Function'.

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

p = figure(title='Sine Wave Function', x_axis_label='X-axis', y_axis_label='Y-axis')

p.line(x, y, line_width=2)

p.grid.grid_line_alpha = 0.8

show(p)
```

```python
# 19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
# values, add hover tooltips to display exact values, label the axes, and set the title as ‘Random Categorical
# Bar Chart'.

output_notebook()

categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))

source = ColumnDataSource(data=dict(categories=categories, values=values))

p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values')

colors = factor_cmap('categories', palette=['#718dbf', '#e84d60', '#ddb7b1', '#c9d9d3', '#80b1d3'], factors=categories)

p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color=colors, legend_field='categories')

hover = HoverTool(tooltips=[('Value', '@values')])
p.add_tools(hover)

p.xaxis.major_label_orientation = 1
p.y_range.start = 0
p.title.align = 'center'
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p, notebook_handle=True)
```
